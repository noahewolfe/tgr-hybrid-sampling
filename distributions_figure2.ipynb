```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from toy import sample_generalized_gaussian

plt.style.use("./tgr-hybrid-sampling.mplstyle")
mpl.rcParams['font.size'] = 15

np.random.seed(1701)
```

```python
fig, ax = plt.subplots()

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$p(x)$')

ax.hist(
    sample_generalized_gaussian(2.5, 5, 2, 10000), 
    density = True,
    bins = 25,
    histtype = 'step',
    label = r'$\gamma = 2$',
    color = '#0072C1'
)
ax.hist(
    sample_generalized_gaussian(2.5, 5, 8, 10000),
    density = True,
    bins = 25,
    histtype = 'step',
    label = r'$\gamma = 8$',
    color = '#FF8C00'
)

ax.legend()

fig.savefig('./figures/toy_model_data.pdf')
```
